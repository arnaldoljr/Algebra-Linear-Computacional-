```python
#AUTOVALORES E AUTOVETORES PELO MÉTODO DA POTÊNCIA INVERSA
def AAInv_mpot(A):   
    l,c = A.shape
    v = np.random.rand(l,1)
    q_zero = v/ LA.norm(v)
    #A = LU 
    LU = linalg.lu_factor(A)
    y_zero = linalg.lu_solve(LU, q_zero) 
    Av_zero = np.dot(q_zero.T,y_zero)
    AvInv_zero = 1/(Av_zero)
    k = 0
    while(k<100):
        q = y_zero /LA.norm(y_zero)
        y = linalg.lu_solve(LU,q)
        AvInv = np.dot(q.T,y)
        #analiso a AV e o AV_zero
        k = k+1
        y_zero = np.copy(y) #autovetor
        AvInv = 1/(AvInv)
        AvInv_zero = np.copy(AvInv) #autovalor
    return (AvInv,y)
```

```python
#Teorema de Gerschgorin
# testes
A = np.array([[10.0,-1.0,0.,1],[0.2,8.0,0.2,0.2],[1.,1.,2.,1.],[-1.,-1.,-1.,-11.,]])
A
```

```python
def Gerschgorin(A):
    l,c = A.shape
    x = sum(np.abs(A[0][:])) - np.abs(A[0][0])
    piso_zero = A[0][0] - x
    teto_zero = A[0][0] + x
    for i in range(1,l):
        x = sum(np.abs(A[i][:])) - np.abs(A[i][i])
        piso = A[i][i] - x
        teto = A[i][i] + x
        if teto>teto_zero:
            teto_zero = teto
        if piso<piso_zero:
            piso_zero = piso
    return(piso_zero,teto_zero)
        
        
    
```

```python
def HH_HENS(A):
    l,c = A.shape
    for i in range(l-2):
        u = A[i+1:,i]
        k = np.linalg.norm(u)
        if u[0]<0.0:
            k = -k  #  PQ DISSO mesmo?
        u[0] = u[0] + k
        h = np.dot(u,u)/2.0 # 1/2 da norma ao quadrado de u 
        v = np.dot(A[i+1:,i+1:],u)/h
        g = np.dot(u,v)/(2*h)
        v = v - g*u
        A[i+1:,i+1:] = A[i+1:,i+1:] - np.outer(v,u)
        A[i,i+1]= -k
    return A
        
        
    
    
```

```python
def TriD_HH(Mat): 
    A = np.copy(Mat)
    l,c = A.shape
    for i in range(l-2):
        x = A[i+1:,i]
        k = np.linalg.norm(x)
        u= np.copy(x)
        if u[0]<0.0:
            k = -k  # PERGUNTAR AO PROFESSOR PQ DISSO
        u[0] = u[0] + k
        h = np.dot(u,u)/2.0 # 1/2 da norma ao quadrado de u 
        Matu = np.outer(u,u)
        n  = len(Matu)
        Q = np.eye(n)
        Q = Q - (Matu/h)
        A_HH = np.eye((n))
        A_HH = A[i+1:,i+1:]
        A_HH = np.dot(np.dot(Q,A_HH),Q)
        Qx = np.dot(Q,x)
        Qx[0] = -k
        A[i,i+1:] = Qx.T
        A[i+1:,i] = Qx
        A[i+1:,i+1:] = A_HH
    return A.round(4)
```

```python
def Diag_QR(A):
    l,c = A.shape
    Qfinal = np.eye((l))
    Q,R = np.linalg.qr(A)
    D = np.dot(R,Q)
    for i in range(c):
        Q,R = np.linalg.qr(D)
        A = np.dot(R,Q)
        Qfinal = np.dot(Qfinal,Q)
        D = np.copy(A)
    return D,Qfinal
    
    
```
